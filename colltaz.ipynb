```python
def collatzstep(n):
    if n%2==0:
        return int((n/2))
    else:
        return 3*n+1
```

```python
def coltz(n):
    p = n
    i = 0
    while p!=1:
        p = collatzstep(p)
        i = i + 1
    return i
```

```python
coltz(10)
```

```python
def findbig(n):
    i = 2
    l = [[2,1]]
    while i <= n:
        if coltz(i) == l[0][1]:
            l.append([i,coltz(i)])
            i = i+1
        elif coltz(i) > l[0][1]:
            l = [[i,coltz(i)]] 
            i = i+1
        else: 
            i=i+1
    return l
        
```
